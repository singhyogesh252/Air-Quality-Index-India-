### Removing the NaNs

##### For AQI Bucket and status (objects)

```python
df_city_day.AQI_Bucket = df_city_day.AQI_Bucket.fillna("Unknown")
df_city_hour.AQI_Bucket = df_city_hour.AQI_Bucket.fillna("Unknown")
df_station_day.AQI_Bucket = df_station_day.AQI_Bucket.fillna("Unknown")
df_station_hour.AQI_Bucket = df_station_hour.AQI_Bucket.fillna("Unknown")
df_stations.Status = df_stations.Status.fillna("Unknown")
```

###### Interpolating

```python
df_city_day.interpolate(limit_direction="both",inplace=True)
df_city_hour.interpolate(limit_direction="both",inplace=True)
df_station_day.interpolate(limit_direction="both",inplace=True)
df_station_hour.interpolate(limit_direction="both",inplace=True)
```

### EDA

###### Checking the total types of stations

```python
sns.countplot(y = "Status", data = df_stations,palette ="viridis")
```

###### Checking the unknown type of stations city wise

```python
unknown_station = df_stations[df_stations['Status']=='Unknown'].sort_values(by='City')
```

```python
plt.figure(figsize=(8,20))
sns.countplot(y = "City" , data = unknown_station,palette ="viridis")
sns.set(style="darkgrid")
plt.plot()
```

##### Top 5 cities with unknown number of stations

### City wise yearly average pollution

##### Ahemdabad

```python
df_ahmedabad = df_city_day[df_city_day.City=="Ahmedabad"]
df_ahmedabad["year"] = pd.DatetimeIndex(df_ahmedabad["Date"]).year
df_ahmedabad["month"] = pd.DatetimeIndex(df_ahmedabad["Date"]).month
```

```python
df_ahmedabad
```

```python
plt.figure(figsize = (14,6))
sns.lineplot(data=df_ahmedabad.groupby("year").mean().drop(['AQI','month'],axis=1),dashes=False)
```
